# Types and Objects

Before progressing to talking about how we use object-oriented programming in Python, we must covere a little theory and make some key definitions. These definitions will underpin the rest of the discussions in this workshop.

In Python, everything is an object. An object is an instance of a "type", which may be thought of as a template for a particular piece of information or group of bits of information.  We may find out the type of an object using the ```type``` function:

````
print(type(3.14))
````

````
a = 2
print(type(a))
````

````
print(type(False))
````

````
print(type(print))
````

````
def sample_function():
  print("Sample")

print(type(sample_function))
````

````
print(type(int))
````

As you can see, not only different values may be examples of different types, but functions, such as the ```print``` function and a user-defined function are also an examples of types.

You may think of a type as a blueprint for a particular arrangement of data, although it does not contain any data. The ```int``` type is a blueprint for the information required to define a whole number, but the variable ```a``` in the code segment above is an object of the ```int``` type with information which represents the value 2.

An object of a certain type is sometimes referred to as an "instance" of the that type.

It is worth noting that the word "class" is almost interchangable with the word "type" in modern Python, although their meanings differed more in earlier versions of the language.
